```python
## Demonstrate built-in data types

values = [10, 3.14, 2+3j, 'Hello', [1,2], (1,2), {1,2}, {'a':1}, True, None]
for v in values:
  print(v, type(v), id(v))
```

```python
# Mutable vs Immutable
x = 10
y = x
x += 5
print(x, y)
lst = [1,2]
a = lst
lst.append(3)
print(lst, a)
```

```python
# Operators
a, b = 10, 3
print(a+b, a-b, a*b, a/b, a%b, a**b, a//b)
print(a>b, a==b)
print(True and False, True or False, not True)
print(a&b, a|b, a^b, a<<1, a>>1)
```

```python
# Dynamic datatype conversion
s = input('Enter value: ')
for t in (int, float):
  try:
    print(t.__name__, t(s))
    break
  except:
      pass
  else:
    print('string', s)
```

```python
# Separate mixed list by datatype
data = [1, 'x', 2.5, [1], True, (1,2)]
result = {}
for item in data:
    key = type(item).__name__
result.setdefault(key, []).append(item)
print(result)
```

```python
# Complex arithmetic
a = 2+3j
b = 1-2j
print('Add:', a+b)
print('Sub:', a-b)
print('Mul:', a*b)
print('Div:', a/b)
```

```python
# Count vowels consonants digits special chars
s = input('Enter string: ')
v=c=d=sp=0
for ch in s:
    if ch.lower() in 'aeiou':
        v+=1
    elif ch.isalpha():
        c+=1
    elif ch.isdigit():
        d+=1
    else:
        sp+=1
print('Vowels',v,'Consonants',c,'Digits',d,'Special',sp)
```

```python
# Palindrome with and without slicing
s = input('Enter string: ')
print('Using slicing:', s == s[::-1])
rev = ''
for ch in s:
    rev = ch + rev
print('Without slicing:', s == rev)
```

```python
# List operations
l = [3,1,2]
l.insert(1,9)
l.append(4)
l.extend([5,6])
l.remove(9)
l.pop()
l.sort()
l.reverse()
print(l)
```

```python
# Second largest and second smallest
nums = [5,1,9,7,2]
first = second = -999999
small = second_small = 999999
for n in nums:
  if n > first:
    second = first
    first = n
  elif n > second and n != first:
    second = n
  if n < small:
    second_small = small
    small = n
  elif n < second_small and n != small:
    second_small = n
print('Second Largest:', second)
print('Second Smallest:', second_small)
```

```python
# Tuple packing and unpacking with nested tuples
t = (1, 2, (3, 4))
a, b, (c, d) = t
print(a, b, c, d)
```

```python
# Remove duplicates while maintaining order
lst = [1,2,1,3,2,4]
res = []
for x in lst:
    if x not in res:
        res.append(x)
print(res)
```

```python
# Dictionary of student marks
marks = {'A':80,'B':95,'C':70}
topper = max(marks, key=marks.get)
avg = sum(marks.values())/len(marks)
sorted_marks = sorted(marks.items(), key=lambda x:x[1], reverse=True)
print('Topper:', topper)
print('Average:', avg)
print(sorted_marks)
```
