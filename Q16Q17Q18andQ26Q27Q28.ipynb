```python
# -*- coding: utf-8 -*-
# Copy file removing digits
"""
Created on Sun Apr 19 15:32:24 2026

@author: asusl
"""

path=r"C:\Users\asusl\Desktop\file_handling\sample1.txt" 
path1=r"C:\Users\asusl\Desktop\file_handling\sample2.txt" 

# Copy file removing digits
with open(path,'w') as f: 
    f.write('ab12cd34')
with open(path) as f: 
    data = f.read()
clean = ''.join(ch for ch in data if not ch.isdigit())
with open(path1,'w') as f: 
    f.write(clean)
print(clean)


```
